### Part 1: Building the base Anagram Checker

Given two valid strings, check to see if they are anagrams of each other. If it is, return `True`, else `False`. For this part, we can assume that uppercase letters are the same as if it was a lowercase character.

Examples of anagrams:
* Silent and Listen
* Night and Thing

Example outputs:
```python
anagram_checker("Silent", "listen") # True
anagram_checker("Silent", "Night") # False
anagram_checker("night", "Thing") # True
```

````python
# For testing purposes, we will write our code in the function
def anagram_checker(word_a, word_b):
  '''Checks if the two words are anagrams of each other.'''
  word_a = word_a.replace(" ", "").lower() #this line removes spaces, converts everything to lowercase and assigns it to word a.
  word_b = word_b.replace(" ", "").lower() #this line removes spaces, converts everything to lowercase and assigns it to word b.
  return sorted(word_a) == sorted(word_b) # if all the letter are the same it should return as true. 
  # My code is now above this line. Testing to see if it works, if not return here and edit it. Hope the grader enjoys my notes along the way.
  #Last note: I understand I didn't need to remove spaces since since we use single words anyways but I thought it would be cool to add.  

# Run your code to check using the words below:
anagram_checker("Silent", "listen")
````

### Part 2: Expanding the functionality of the Anagram Checker

Using your existing and functional anagram checker, let's add a boolean option called `is_case_sensitive`, which will return `True` or `False` based on if the two compared words are anagrams and if we are checking for case sensitivity.

````python
def anagram_checker(word_a, word_b, is_case_sensitive):
  '''This anagram checker allows the user to specify whether or not the anagram should be case sensitive, using true or false in the last parameter.'''
  word_a = word_a.replace(" ", "") #this line removes spaces and assigns it to word a.
  word_b = word_b.replace(" ", "") #this line removes spaces and assigns it to word b. 
  
  if is_case_sensitive == True:
    return sorted(word_a) == sorted(word_b) # if all the letter are the same it should return as true. 
  elif is_case_sensitive == False:
    word_a = word_a.lower() #this line converts everything to lowercase and assigns it to word a.
    word_b = word_b.lower() #this line converts everything to lowercase and assigns it to word b.
    return sorted(word_a) == sorted(word_b) # if all the letter are the same it should return as true. 
  else: 
    return "Please enter true or False for the last parameter." #I added this as a catch if the case sensitivity parameter was not filled in.
  # Not sure if adding the else statement is necessary for this assignment or if it is even tested, but I was playing around and thought it would
  # be neat to add in! 
  
  # Again all my code is above this line, and I think I explained each part pretty well (?). Hopefully they don't make this to messy to read.

# Run your code to check using the words below:
anagram_checker("Silent", "listen", False) # True
````
